9.2 Define a generator function called get_odds() that returns the odd numbers from range(10). Use a for loop to find and print the third value returned.

```python
def get_odds() -> list[int]:
    odds_list = []
    for i in range(10):
        if i%2 == 1:
            odds_list.append(i)
    return odds_list
odds = get_odds()
print(odds[2])
```
