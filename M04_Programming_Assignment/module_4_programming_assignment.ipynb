```python
import sqlite3
conn = sqlite3.connect("books.db")
curs = conn.cursor()
curs.execute("CREATE TABLE IF NOT EXISTS books (title VARCHAR(20) PRIMARY KEY, author VARCHAR(20), year INT)")
conn.commit()
conn.close()
```

```python
import sqlite3
import csv

conn = sqlite3.connect("books.db")
curs = conn.cursor()

with open("books2.csv", "r") as f:
    f_text = csv.reader(f)
    header = next(f_text)
    for row in f_text:
        curs.execute("INSERT INTO books (title, author, year) VALUES (?, ?, ?)", (row[0], row[1], int(row[2])))
        print(row[0] + " added successfully.")
conn.commit()
conn.close()
```

```python
from sqlalchemy import create_engine, text, select

engine = create_engine("sqlite:///books.db")

with engine.connect() as conn:
    rows = conn.execute(text("SELECT title FROM books ORDER BY title ASC"))
    for row in rows:
        print(row)
```
